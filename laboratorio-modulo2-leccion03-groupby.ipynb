```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
netflix_original = pd.read_csv("datos/netflix_originals.csv")
netflix_original.drop(columns = ["Unnamed: 0"], inplace=True)
netflix_original
```

```python
netflix_titles =pd.read_csv("datos/netflix_titles.csv")
netflix_titles.drop(columns = ["show_id","Unnamed: 0"], inplace=True)
netflix_titles
```

```python
netflix_original.rename(columns={'Title': 'title'}, inplace=True)

df= pd.merge(netflix_titles, netflix_original, on=["title"], how='outer')

df.head()
```

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
añoactual= pd.Timestamp.now().year
df_ultimos_5_años = df[df['release_year']>=2019]

# Agrupar por país e idioma, y contar el número de títulos en cada combinación
conteo_idioma = df_ultimos_5_años.groupby(['country', 'Language']).size().reset_index(name='conteo')

# Obtener el idioma más frecuente por país
idioma_mas_frecuente_por_pais = conteo_idioma.loc[conteo_idioma.groupby('country')['conteo'].idxmax()]
idioma_mas_frecuente_por_pais.sort_values('conteo',ascending=False)
```
