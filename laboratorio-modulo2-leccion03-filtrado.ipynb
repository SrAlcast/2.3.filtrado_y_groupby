# Laboratorio Filtrado de Datos

```python
# Carga el dataset de trabajo
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
netflix_original = pd.read_csv("datos/netflix_originals.csv")
netflix_original.drop(columns = ["Unnamed: 0"], inplace=True)
netflix_original
```

```python
netflix_titles =pd.read_csv("datos/netflix_titles.csv")
netflix_titles.drop(columns = ["show_id","Unnamed: 0"], inplace=True)
netflix_titles
```

```python
netflix_original.rename(columns={'Title': 'title'}, inplace=True)
df= pd.merge(netflix_titles, netflix_original, on=["title"], how='outer')
df.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_movie=df["type"]=="Movie"
condicion_2021=df["release_year"]==2021

df_2021_Movie=df[condicion_2021&condicion_2021]
df_2021_Movie.head()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
condicion_90m=df["duration"]=="90 min"

df_90m=df[condicion_90m].filter(["title","duration"])
df_90m.head()
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
condicion_india=df["country"].str.contains('India',na=False)
condicion_a2018=df['release_year']>2018

df_india_a2018=df[condicion_india&condicion_a2018]
df_india_a2018[['title','country','release_year']]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_TV14=df["rating"]=='TV-14'
condicion_R=df['rating']=='R'

df_TV14_R=df[condicion_TV14|condicion_R]
df_TV14_R
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
condicion_Christopher=df["director"]=='Christopher Nolan'
condicion_Martin=df['director']=='Martin Scorsese'

df_Christopher_Martin=df[condicion_Martin|condicion_Christopher]
df_Christopher_Martin.shape
df_Christopher_Martin.shape
print(f"Dirigidas por Christopher o Martin:{df_Christopher_Martin.shape[0]}")
print(f"Dirigidas por Christopher:{df[condicion_Christopher].shape[0]}")
print(f"Dirigidas por Martin:{df[condicion_Martin].shape[0]}")
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
condicion_EEUU=df["country"].str.contains('United States',na=False)
condicion_Canada=df["country"].str.contains('Canada',na=False)
condicion_UK=df["country"].str.contains('United Kingdom',na=False)
df_1=df[condicion_EEUU|condicion_Canada|condicion_UK]
condicion_language=df_1['Language'].str.contains('/',na=False)
df_language=df_1[condicion_language]
print(f'Peliculas con más de un lenguage:{df_language.shape[0]}')
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
titles2015_2020=df['release_year'].between(2015,2020, inclusive='both')
df[titles2015_2020]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_imd=df['IMDB Score'].between(6.5,8.0, inclusive='both')
df[condicion_imd]['IMDB Score']
#df[condicion_imd]['IMDB Score'].unique()
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
condicion_90m=df['duration']=='90 min'
condicion_150m=df['duration']=='150 min'
df[condicion_90m|condicion_150m]
#df[condicion_90m|condicion_150m]['duration'].unique()
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
condiciontvshow=df['type']=='TV Show'
condicionLove=df['title'].str.contains('Love', na=False)
df[condicionLove&condiciontvshow]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
condicion1=df['type'].str.contains('Movie', na=False)
condicion2=df['release_year']<2010
print(df[condicion1&condicion2].shape[0])
print(df[condicion1&condicion2]['type'].unique())
print(df[condicion1&condicion2]['release_year'].unique())
df[condicion1&condicion2]
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
df_top10 = df.nlargest(10, 'IMDB Score').filter(['title', 'release_year', 'IMDB Score'])
df_top10
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
df_top5 = df.nsmallest(5, 'IMDB Score').filter(['title', 'release_year', 'IMDB Score'])
df_top5.sort_values(by='IMDB Score', ascending=False)
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
condicion1=df['title'].notnull()
condicion2=df['country'].notnull()
condicion3=df['IMDB Score'].notnull()
print(df[condicion1&condicion2&condicion3].shape[0])
print(df[condicion1&condicion2&condicion3]['title'].unique())
print(df[condicion1&condicion2&condicion3]['country'].unique())
df[condicion1&condicion2&condicion3]
```
